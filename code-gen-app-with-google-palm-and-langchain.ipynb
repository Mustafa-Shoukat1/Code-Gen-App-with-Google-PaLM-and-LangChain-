<h1 style="color: #6a0dad; font-family: 'Times New Roman', Times, serif;">Define a prompt template for code generation </h1>

```python
from langchain.prompts import PromptTemplate
# Define a prompt template for code generation
prompt_template = PromptTemplate(
    input_variables=["task"],
    template="Generate a {task} in Python.",
)
prompt_template
```

<h1 style="color: #6a0dad; font-family: 'Times New Roman', Times, serif;">Define a Function code generation </h1>

```python
def generate_code(task):
    prompt = prompt_template.format(task=task)
    return llm.invoke(prompt)
```

<h1 style="color: #6a0dad; font-family: 'Times New Roman', Times, serif;">User Queries for code generation </h1>

```python
tasks = [
    "function that calculates the factorial of a number",
    "function to find the maximum element in an array",
    "give me code for complete application of  langchain.llms with GooglePalm code generation application  ",
    "give me code to load SQL in with python in jupter notebook",
    "give me code to load nlp reviews classification model for hugging face"
]
```

<h1 style="color: #6a0dad; font-family: 'Times New Roman', Times, serif;">Generate code for each task </h1>

```python
for task in tasks:
    print(f"Task: {task}\n")
    print(f"Generated Code:\n{generate_code(task)}\n")
    print("="*50 + "\n")
```

<h1 style="color: #6a0dad; font-family: 'Times New Roman', Times, serif;">Define Complex Query for code Generation</h1>

```python
print(f"Generated Code:\n{generate_code('''Create dummy data and generate code to load a sentiment analysis model from the Hugging Face pipeline using a simple model. Provide the complete code for a Kaggle kernel. ''')}\n")
print("="*50 + "\n")
```
